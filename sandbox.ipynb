```python
import os
```

```python
folder = "txt_output/trong_v11m_v2data_5_classes"   # đổi lại đường dẫn của bạn

for fname in os.listdir(folder):
    if fname.endswith(".txt"):
        path = os.path.join(folder, fname)

        new_lines = []
        with open(path, "r") as f:
            for line in f:
                items = line.strip().split()  # tách theo mọi loại whitespace

                if len(items) != 6:
                    continue

                id_, conf, x1, y1, x2, y2 = items
                new_line = " ".join([id_, x1, y1, x2, y2, conf])
                new_lines.append(new_line)

        with open(path, "w") as f:
            f.write("\n".join(new_lines))

print("Done!")
```
